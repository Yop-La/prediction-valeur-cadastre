```python
import pandas as pd
```

### Import des pop ens

```python
years = ["2014","2015"]
```

```python
data_pop = {}
for year in years:
    pop = pd.read_csv('../../data/csp/ensemble_pop_'+ year + '.csv',sep=',',dtype= {'code_commune' : str})
    # pop.rename(columns = {'population_totale' : 'pop_' + year},inplace=True)
    data_pop[year] = pop
```

### Import de immo by year

```python
immo = pd.read_csv('../../data/join/immo_by_year.csv',sep=',',dtype= {'code_commune' : str})
```

```python
immo.head()
```

### Join 

```python
immo_pop.shape
```

```python
immo_pop = immo
for year in years:
    immo_pop = pd.merge(immo_pop, data_pop[year], on='code_commune', how='inner')
```

```python
immo_pop.to_csv('../../data/join/join_immo_pop_ens.csv',index=False)
```
