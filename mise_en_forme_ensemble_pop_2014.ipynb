### On renomme les colonnes 

```python
pop2014 = pop2014.rename(columns={'COM': 'code_commune'})
```

```python
pop2014.columns = pop2014.columns.map(lambda x : x.lower())
```

```python
pop2014.head()
```

### On groupe par code commune

```python
pop2014commune = pop2014.groupby('code_commune').sum().reset_index()
```
