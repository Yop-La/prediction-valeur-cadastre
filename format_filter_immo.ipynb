### Enlever les lignes où le nombre de mutations est inférieure à 10 toute années confondues

```python
cols = immo.columns.values
col_to_filer = "nb_mutation"
```

```python
filtre = True
for col in cols:
    start_with = col.startswith(col_to_filer)
    if start_with:

        filtre = filtre & np.array(list(immo[col] >= 10))
immo_filtrer = immo[filtre]
```

```python
immo_filtrer.sort_values(by=['nb_mutation_2015']).head()
```

### Tronquer les valeurs

```python
col_types = immo_filtrer.dtypes
float_filter = immo_filtrer.dtypes == "float64"
```

```python
immo_filtrer.loc[:,float_filter] = immo_filtrer.loc[:,float_filter].astype(int)
```

```python
immo_filtrer.to_csv('../../data/join/immo_filtrer.csv',index=False)
```
