```python
import pandas as pd
```

### Import des pop

```python
years = ["2014","2015","2016"]
```

```python
data_pop = {}
for year in years:
    pop = pd.read_csv('../../data/population/pop_commune_'+ year + '_clean.csv',sep=',',dtype= {'code_commune' : str})
    pop.rename(columns = {'population_totale' : 'pop_' + year},inplace=True)
    data_pop[year] = pop
```

### Import de immo by year

```python
immo = pd.read_csv('../../data/join/immo_filtrer.csv',sep=',',dtype= {'code_commune' : str})
```

```python
immo.head()
```

### Join 

```python
data_pop["2016"].head()
```

```python
immo.shape
```

```python
immo_pop = immo
for year in years:
    immo_pop = pd.merge(immo_pop, data_pop[year], on='code_commune', how='inner')
```

```python
immo_pop.to_csv('../../data/join/join_immo_pop.csv',index=False)
```
