```python
import pandas as pd
import numpy as np
```

```python
immo = pd.read_csv('../../data/join/immo_by_month_100.csv',sep=',',dtype= {'code_commune' : str}).rename(columns = {'id_bien_int':'id_bien'})
```

```python
immo.shape
```

```python
immo.head(13)
```

### Feature enginneering sur les timesetps ( month + year )

```python
immo['mnth_sin'] = np.sin((immo.month-1)*(2.*np.pi/12))
immo['mnth_cos'] = np.cos((immo.month-1)*(2.*np.pi/12))
```

```python
immo['mois'] = immo['month'] + (immo['year']-2014)*12
```

```python
immo.head(12)
```

```python
immo2 = immo.copy()
```

### Feature enginneering - ajout des variables observées à  t-i  ( lagk) avec i  dans [[1;k]]

```python
cols_to_extract = ['valeur']
```

```python
k = 12
```

```python
for i in range(1,k+1):
    lag1 = list(map(lambda x: x + "-"+str(i),cols_to_extract))
    immo2[lag1] = immo2.groupby(['id_bien'])[cols_to_extract].shift(i)
```

```python
immo2.sort_values(by=['id_bien','mois']).head(20)
```

### Feature enginneering - ajout des différences entre t-i et t-(i+1) (diffi) pour i dans [[1;k]]

```python
k=10
for i in range(1,k+1):
    lag1 = list(map(lambda x: x + "-"+str(i),cols_to_extract))
    diff1 = list(map(lambda x: x + "_diff-"+str(i)+"-"+str(i+1),cols_to_extract))
    immo2[diff1] = immo2.groupby(['id_bien'])[lag1].diff()
```

```python
immo2.sort_values(by=['id_bien','mois']).head(20)
```

```python
immo2.shape
```

```python
immo_train = immo2.dropna()
```

```python
immo_train.shape
```

### Export

```python
immo_train.to_csv('../../data/ml/train_immo_100_more_k_12_10.csv',index=False)
```
