```python
def get_lbc_immo_ads(city, label_city, offset):
    offset = str(offset)
    headers = {
        'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:67.0) Gecko/20100101 Firefox/67.0',
        'Accept': '*/*',
        'Accept-Language': 'en-US,en;q=0.5',
        'Referer': 'https://www.leboncoin.fr/recherche/?category=9&locations=Rennes&page=2',
        'api_key': 'ba0c2dad52b3ec',
        'Content-Type': 'text/plain;charset=UTF-8',
        'Origin': 'https://www.leboncoin.fr',
        'Connection': 'keep-alive',
    }

    data = '{"filters":{"category":{"id":"9"},"enums":{"ad_type":["offer"]},"keywords":{},"location":{"locations":[{"locationType":"city","city":"'+city+'","label":"'+label_city+'"}]},"ranges":{}},"limit":35,"limit_alu":3,"offset":'+offset+'}'
    response = requests.post('https://api.leboncoin.fr/finder/search', headers=headers, data=data)
    
    body = response.content
    body = json.loads(body)
    
    return(body)
```

```python
def get_lbc_city(commune):
    
    commune = str(commune)
    
    headers = {
        'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:67.0) Gecko/20100101 Firefox/67.0',
        'Accept': '*/*',
        'Accept-Language': 'en-US,en;q=0.5',
        'Referer': 'https://www.leboncoin.fr/recherche/?category=9&locations='+commune+'&page=2',
        'Content-Type': 'text/plain;charset=UTF-8',
        'Origin': 'https://www.leboncoin.fr',
        'Connection': 'keep-alive',
    }

    data = '{"text":"ly","context":{}}'

    response = requests.post('https://api.leboncoin.fr/api/parrot/v2/cityzipcode/complete', headers=headers, data=data)

    body = response.content
    body = json.loads(body)
    
    return(body[0])
```
