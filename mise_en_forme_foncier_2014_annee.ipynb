## On enlève les duplicates

####  avec une règle maison ( on prend le premier élement des groupes fait sur l'id de mutation )

```python
nb_id_mutation = immo2014.groupby(['id_mutation']).size()
nonunique_id_mutation = nb_id_mutation[nb_id_mutation != 1]
nonunique_id_mutation = nonunique_id_mutation.index
```

```python
immo2014[immo2014.id_mutation.isin(nonunique_id_mutation)].sort_values(by=['id_mutation'], ascending=False)
```

```python
immo2014 = immo2014.groupby('id_mutation').first().sort_values(by=['id_mutation'], ascending=False)
```

```python
immo2014.shape
```

```python
immo2014.isna().sum()
```

### Group by (  code_commune, nom_commune , type_local, nombre_pieces_principales )

```python
immoGroup2014 = immo2014
```

```python
immoGroup2014.reset_index( inplace=True)
```

```python
immoGroup2014.head()
```

```python
cols_to_keep = ["code_commune","nom_commune","code_type_local", "type_local","nombre_pieces_principales","surface_reelle_bati","valeur_fonciere","id_mutation"]
```

```python
immoGroup2014 = immoGroup2014[cols_to_keep]
```

```python
immoGroup2014 = immoGroup2014.groupby(["code_commune","nom_commune","type_local","nombre_pieces_principales"]).agg({'surface_reelle_bati':'mean', 'valeur_fonciere':'mean','id_mutation':'size'}).rename(columns={'id_mutation':'nb_mutation'}).reset_index()
```

```python
immoGroup2014.head()
```

### Création d'un identifiant de bien

```python
immoGroup2014['id_bien'] = immoGroup2014["code_commune"] + "_" + immoGroup2014["code_type_local"].astype(int).astype(str) + "_" + immoGroup2014["nombre_pieces_principales"].astype(int).astype(str)
```

```python
immoGroup2014.to_csv('../data/immobilier_2014_clean.csv',index=False)
```
